# Anaconda

In this lesson, you will be getting a quick glimpse at the Anaconda environment - one of the most popular environments for doing data analysis in Python.

[Anaconda](https://anaconda.org/) is a distribution of packages built for data science. It comes with conda, a package and environment manager. You'll be using conda to create environments for isolating your projects that use different versions of Python and/or different packages. You'll also use it to install, uninstall, and update packages in your environments. Using Anaconda has made my life working with data much more pleasant.

# Installing Anaconda

Anaconda is available for Windows, Mac OS X, and Linux. You can find the installers at https://www.anaconda.com/download/ and the installation instructions [here](https://docs.anaconda.com/anaconda/install/).

# Managing Packages

## Install
* To install a package `conda install package_name`
 * Example `conda install numpy`
 
* Install multiple packages at the same time `conda install package_name_1 package_name_2 package_name_3`
 * Example `conda install numpy scipy pandas`
 
* To install a specific package version `conda install package_name=<version>`
 * Example `conda install numpy=1.10`
 
## Uninstall
* To remove a package `conda remove package_name`
 * Example `conda remove numpy`
 
## Update
* To update a package `conda update package_name`
 * Example `conda update numpy`
 
* Update all environment `conda update --all`

## List
* List installed packages `conda list`

## Search
* If you don't know the exact name of the package you're looking for, use `conda search *search_term*`
 * Example `conda search '*beautifulsoup*'`

# Managing environments

## Create environments
* create an environment use `conda create -n env_name list of packages`
 * Example `conda create -n my_env numpy`
* Is useful create both Python 2.x and Python 3.x environment for any specific project
 * `conda create -n py2 python=2`
 * `conda create -n py3 python=3`

## Entering and leaving an environment
* To enter use `conda activate my_env`
* To exit use `conda deactivate`

## Saving and loading environments
* You can save the packages to a YAML file `conda env export > environment.yaml`
* You can create an environment using YAML file `conda env create -f environment.yaml`

## List environments
* To list use `conda env list`

## Remove environments
* To remove use `conda env remove -n env_name`

# General notes

## Anaconda
* **Anaconda** has simplified my workflow and solved a lot of issues I had dealing with packages and multiple Python versions

* If you don't need all the packages or need to conserve bandwidth or storage space, there is also **Miniconda**

* **Conda** is a program you'll be using exclusively from the command line
 * use conda to manage your packages and environments (like **pip**)
 
## Environments
* Conda is also a virtual environment manager (like virtualenv and pyenv)
 * you can run different versions of packages or languages (like Python 2 and Python 3 in same machine)
  
* You can also export the list of packages in an environment to a file
 * similar to `pip freeze > requirements.txt`

* Command to entering and leaving an environment can differs among operation systems
