# Data Analysis Process Overview

## Step 1: Ask questions
Either you're given data and ask questions based on it, or you ask questions first and gather data based on that later. In both cases, great questions help you focus on relevant parts of your data and direct your analysis towards meaningful insights.

## Step 2: Wrangle data
You get the data you need in a form you can work with in three steps: gather, assess, clean. You gather the data you need to answer your questions, assess your data to identify any problems in your data’s quality or structure, and clean your data by modifying, replacing, or removing data to ensure that your dataset is of the highest quality and as well-structured as possible.

## Step 3: Perform EDA (Exploratory Data Analysis)
You explore and then augment your data to maximize the potential of your analyses, visualizations, and models. Exploring involves finding patterns in your data, visualizing relationships in your data, and building intuition about what you’re working with. After exploring, you can do things like remove outliers and create better features from your data, also known as feature engineering.

## Step 4: Draw conclusions (or even make predictions)
This step is typically approached with machine learning or inferential statistics that are beyond the scope of this course, which will focus on drawing conclusions with descriptive statistics.

## Step 5: Communicate your results
There are many ways to communicate your results: reports, slide decks, blog posts, emails, presentations, or even conversations. Data visualization will always be very valuable.

## Headers
Another thing you can do with `read_csv` is specify which line of the file is the header, which specifies the column labels. It's usually the first line, but sometimes we'll want to specify a later line if there is extra meta information at the top of the file. We can do that like this.

```python
df = pd.read_csv('./data/student_scores.csv', header=2)
df.head()
```

Here, row 2 was used as the the header and everything above that was cut off. By default, `read_csv` uses header=0, which uses the first line for column labels.

If columns labels are not included in your file, you can use `header=None` to prevent your first line of data from being misinterpreted as column labels.

```python
df = pd.read_csv('./data/student_scores.csv', header=None)
df.head()
```

You can also specify your own column labels like this.

````python
labels = ['id', 'name', 'attendance', 'hw', 'test1', 'project1', 'test2', 'project2', 'final']
df = pd.read_csv('./data/student_scores.csv', header=0, names=labels)
df.head()
````

# Drawing Conclusions Example

Let's use descriptive statistics and visualizations to address the question we posed with cancer data.

**Does the size of tumor affect its malignancy?**

We had compare the size of benign tumors vs. malignant tumors. We can use masks to select all rows that were diagnosed as malignant and benign, put them in separate dataframes.

First, create a new dataframe called `df_m` that contains only malignant tumors.

````python
import pandas as pd
df = pd.read_csv('./data/cancer_data_edited.csv')
df.head()
````

````python
df_m = df[df['diagnosis'] == 'M']
df_m.head()
````

The part `df['diagnosis'] == 'M'` make pandas return a series of booleans indicating whether the value in the diagnosis columns is equal to `M`.

````python
mask = df['diagnosis'] == 'M'
print(mask)
````

Indexing the original dataframe with this mask will return all rows where the value in mask is `True`. In other words, where the diagnosis is `M`.

Now see the summary statistics about the area feature which offers a good metric for size.

````python
df_m['area'].describe()
````

Do the same for all benign tumors

````python
df_b = df[df['diagnosis'] == 'B']
df_b.head()
````

````python
df_b.describe()
````

````python
print('The mean area of malignant tumors is {} while that of benign tumors is {}'.format(df_m['area'].mean(), df_b['area'].mean()))
````

Summary statistics like the mean are helpful, it would be nice to be able to compare the distibutions of the areas of both tumors visually. 

Don't worry about understanding this code righ now.

````python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(df_b['area'], alpha=0.5, label='benign')
ax.hist(df_m['area'], alpha=0.5, label='malignant')
ax.set_title('Distributions of Benign and Malignant Tumors Areas')
ax.set_xlabel('Area')
ax.set_ylabel('Count')
ax.legend(loc='upper right')
plt.show()
````

This visual suggests that there is a difference between the distribution of areas for benign and malignant tumors. We do not yet have the tools to conclude that these two distributions are different or whether the size definitely affects the tumor's malignancy.

However, we can observe from summary statistics and these histograms that malignant tumors are generaly larger in size than benign tumors.

# Communicating Results Example

Create two dataframes to serparate people who make above and below $50000.

````python
import pandas as pd
%matplotlib inline

df_census = pd.read_csv('./data/census_income_data.csv')

df_a = df_census[df_census['income'] == ' >50K']
df_b = df_census[df_census['income'] == ' <=50K']
````

We can use bar graphs to compare the education levels reached in both groups.

````python
df_a['education'].value_counts().plot(kind='bar');
````

````python
df_b['education'].value_counts().plot(kind='bar');
````

Here we don't have consistent x-axis labels for these graphs. It's ideal to have the labels in the same order if we want to make a clear comparison.

Let's order both results from value counts with the same index.

````python
ind = df_a['education'].value_counts().index
````

````python
df_a['education'].value_counts()[ind].plot(kind='bar');
````

````python
df_b['education'].value_counts()[ind].plot(kind='bar');
````

Next, let's plot pie charts to compare what work classes dominate in each group.

````python
ind = df_a['workclass'].value_counts().index
df_a['workclass'].value_counts()[ind].plot(kind='pie', figsize=(8, 8));
````

Lastly, let's use histograms to plot the distribuition of ages for each group.



````python
df_a['age'].hist();
````

````python
df_b['age'].hist();
````

Interesting, the people who make more than 50000 seem to be older.

````python
df_a['age'].describe()
````

````python
df_b['age'].describe()
````

If we look at the summary statistics, we can see that the mean age for those making less is about 37, and 44 for those making more.

This shows us that people making more are generally older but these visualizations give us a much better idea of how those ages are distributed in these examples.
